```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load and Preprocess Data
# Reads the 561-feature vectors and labels for training and testing.

# Applies label_map to convert numeric labels (1â€“6) into human-readable activity names.

# Scales the feature vectors using StandardScaler to normalize data (mean=0, std=1).
```

```python
# Load data
X_train = pd.read_csv('train/X_train_data.txt', sep='\s+', header=None)
y_train = pd.read_csv('train/y_train_data.txt', header=None)
X_test = pd.read_csv('test/X_test_data.txt', sep='\s+', header=None)
y_test = pd.read_csv('test/y_test_data.txt', header=None)

# Map numeric labels to activity names
label_map = {
    1: "Walking", 2: "Walking_Upstairs", 3: "Walking_Downstairs",
    4: "Sitting", 5: "Standing", 6: "Laying"
}
y_train = y_train[0].map(label_map)
y_test = y_test[0].map(label_map)

# Normalize
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Define the Evaluation Function

# A reusable function to:

# Print classification metrics: precision, recall, F1-score, accuracy

# Plot a confusion matrix using Seaborn for visual comparison of predicted vs actual activities
```

```python
def evaluate_model(y_true, y_pred, model_name):
    print(f"=== {model_name} ===")
    print(classification_report(y_true, y_pred))
    
    cm = confusion_matrix(y_true, y_pred, labels=list(label_map.values()))
    plt.figure(figsize=(8,6))
    sns.heatmap(cm, annot=True, fmt='d', cmap="Blues",
                xticklabels=label_map.values(),
                yticklabels=label_map.values())
    plt.title(f"{model_name} Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.show()
```

```python
# Train & Evaluate Support Vector Machine (SVM)

# Trains an SVM classifier using the Radial Basis Function (RBF) kernel.

# Predicts on the test set and evaluates performance using the custom function.


```

```python
from sklearn.svm import SVC
svm = SVC(kernel='rbf')
svm.fit(X_train_scaled, y_train)
y_pred_svm = svm.predict(X_test_scaled)
evaluate_model(y_test, y_pred_svm, "Support Vector Machine")
```

```python
# Train & Evaluate Random Forest

# Trains a Random Forest classifier with 100 trees.

# Evaluates test performance with classification metrics and a confusion matrix.
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train_scaled, y_train)
y_pred_rf = rf.predict(X_test_scaled)
evaluate_model(y_test, y_pred_rf, "Random Forest")
```

```python
# Train & Evaluate K-Nearest Neighbors (KNN)

# Trains a KNN classifier with k=5.

# Compares model performance visually and numerically.
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)
y_pred_knn = knn.predict(X_test_scaled)
evaluate_model(y_test, y_pred_knn, "K-Nearest Neighbors")
```

```python
# Train & Evaluate Logistic Regression

# Trains a logistic regression model for multi-class classification.

# Evaluates model output using your custom evaluation function.
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=500)
lr.fit(X_train_scaled, y_train)
y_pred_lr = lr.predict(X_test_scaled)
evaluate_model(y_test, y_pred_lr, "Logistic Regression")
```
